```python
crime_2021 = crime[crime["year"] == 2021]
```

```python
crime_2021_10000 = crime_2021.sample(10000)
```

```python
geolocator = geopy.Nominatim(user_agent = "geo2zip")
```

```python
# crime_2021_10000["zipcodes"] = crime_2021_10000.apply(get_zipcode, axis = 1, geolocator = geolocator, 
#                                                       lat_field = "lat", lon_field = "lon")
```

```python
def get_zipcode(geolocator, lat_field, lon_field):
    location = geolocator.reverse((lat_field, lon_field))
    try:
        result = location.raw["address"]["postcode"]
    except:
        result = "00000"
    return result
```

```python
crime_2021_10000["zipcodes"] = crime_2021_10000.apply(lambda x: get_zipcode(geolocator, x.lat, x.lon), axis = 1)
```
